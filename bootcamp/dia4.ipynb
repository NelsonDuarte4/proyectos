```python
contacto = {}
galeria = ["nombre","apellido","numero"]
for clave in galeria:                           # se rellena la lista con un for(ciclo)
    dato = input(f"ingrese el {clave} ")
    contacto[clave] = dato
    

while True :
    opcion = input("modificar, eliminar o salir")
    if opcion == "modificar":
        for clave in contacto:
         print(clave)
        campo = input("que campo quiere modificar")
       
        if campo in contacto:
            nueva_clave = input("ingrese un nuevo dato")
            contacto[campo] = nueva_clave
        else:
            print("esa clave no existe")    

    elif opcion == "eliminar":
    
        for clave in contacto:
            print(clave)

    campo = input("Qué campo quiere eliminar?: ")

    if campo in contacto:
        del contacto[campo]
        print(f"Campo {campo} eliminado correctamente.")
    
    elif opcion == "salir":
        break
    else:
        print("Esa clave no existe.")
        

    
```

```python
class animalito:
    animal = "ave"

    def __init__ (self,volar, planear, posicion = 0):
        self.volar = volar
        self.planear = planear 
        self.posicion = posicion
    
    def aire (self, metros):
        self.posicion += metros
        print(f"volo {metros} m de distancia")
    def caida (self, tiempo):
        print(f"planeo por {tiempo} en el aire" )
    
    pajaro = animalito("")

animalito.aire(70)
```

```python
class Perro:
    tipo = "Perro"
    def init(self, raza, posicion=0):
       self.raza = raza 
       self. posicion = posicion

    def corre (self, corre):
           self.posicion += corre
           print(f"{self.raza}  corre en el parque  { corre}m. Busca su pelota: {self.posicion}m.")

    def ladra(self,ladra):
           print(f"{self.raza} gira y  ladra {ladra}.")


mi_perro = Perro("Delmer")
mi_perro.corre(100)
mi_perro.ladra("Gua gua")
```
