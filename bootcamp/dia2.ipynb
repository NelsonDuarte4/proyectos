```python
#se ingresa los datos
nom_producto = str(input("ingrese el nombre del producto"))
cantidad = int(input("ingrese la cantidad del producto"))
precio = int(input("ingrese el precio de la unidad"))

def stock(nom_producto, cantidad, precio): #Defini la variable con sus parametros
   total = precio * cantidad #calculo del total de los productos
   
   print("RESUMEN DE LAS CARACTERISTICAS")
   print("Nombre del articulo:",nom_producto)
   print("Cantidad de articulos:",cantidad) #imprimo los datos
   print("Precio por unidad:",precio)
   print("El monto total del conjunto de los productos es:",total)

stock(nom_producto, cantidad, precio) 
```
